```
Package folder:
│
├── package_name
│    └── module1.py
│    └── module2.py
│    └── module3.py
│    └── README.md
│    └── license.txt
│    └── __init__.py
│    └── setup.cfg
│
└── setup.py
```

## Troubleshooting Possible Errors

- Two websites: Note that there is a test.pypi website and a pypi website
- Unique Name: Name of the package must be unique.
- Reuploading and versioning: You cannot upload the same version of a package again. 
 - simply change the version number of your package, i.e. `version=0.1.1` to `version=0.1.2`is sufficient.
 
What these version numbers mean, see this tutorial about [semantic versioning](https://semver.org/) --> Standard for deciding when to move up to the next number in your versioning scheme.
